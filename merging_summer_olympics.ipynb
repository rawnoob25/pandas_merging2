*Let's now examine the tsv file that contains data representing all medalists in Editions of the Summer Olympics between 1896 and 2008- inclusive of both*

```python
fn ='Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - ALL MEDALISTS.tsv'
medalists = pd.read_csv(fn, sep='\t', header = 4)
medalists.head()
```

*We'll now split the data frame medalists into a list of dataframes- wherein each data frame corresponds to a particular Summer olympic edition*

<b><em> Note that this is precisely the behavior of the 'split' function in R. The list comprehension in the below cell mimics the split function in R and the factor on which we're splitting is 'Edition'</em></b>

```python
medalists['Edition'].value_counts().sum()
```

```python
#list(medalists.groupby('Edition'))[0]
medalists_by_edition = [grp for ed, grp in medalists.groupby('Edition')]
medalists_by_edition[0].info()
```

```python
medalists_by_edition[0].head()
```

<em>Now, for each data frame in the list medalists_by_edition, let's only keep the columns <b>'Athlete'</b>, <b>'NOC'</b>, <b>'Medal'</b>, and <b>'Edition'</b>.</em>

```python
medalists_by_edition_colsOfInterest = [x[['Athlete','NOC','Medal','Edition']] for x in medalists_by_edition]
medalists_by_edition_colsOfInterest[0].head()
```

*Now let's vertically concatenate these data frames so that we have the 'Athlete', 'NOC', 'Medal', and 'Edition' data for all editions of the summer olympics (between 1896 and 2008- inclusive of both) captured in a single data frame.*

<b><em> Note that we pass ignore_index = True to the concat function since a subset of each index (for each data frame in the list all_medalists_relevantCols) is shared across all the data frames). For some reason that I'm not aware of, not failing to pass ignore_index = True doesn't seem to change the output in this case (see 3 cells below and 4  cells below).</em></b>

```python
all_medalists_relevantCols = pd.concat(medalists_by_edition_colsOfInterest, ignore_index = True)
all_medalists_relevantCols.head()
```

```python
all_medalists_relevantCols.tail()
```

```python
pd.concat(medalists_by_edition_colsOfInterest).head()
```

```python
pd.concat(medalists_by_edition_colsOfInterest).tail()
```

```python
len(all_medalists_relevantCols['NOC'].unique())
```

*We'll now make a 2-way contingency table of medal counts by olympic edition (the index) and country (the columns). Note that this table will have 138 columns since the the 'NOC' field of the data frame all_medalists_relevantCols has 138 unique levels. Using 'Athlete' as the values and 'count' as the aggregation function will give us the table we want.*

```python
medal_counts = all_medalists_relevantCols.pivot_table(index = 'Edition', columns = 'NOC', values = 'Athlete', aggfunc = 'count')
medal_counts.head()
```

```python
medal_counts.tail()
```

<em>We can see that there are a lot of NaN values. This is because many countries won no medals in a given edition of the summer olympics. <b>So, in this instance, NaN actually represents zero.</b> Also these are <b> medal counts</b>, so they're implicitly integer values.  We'll change medal_counts to reflect these two facts.</em>

```python
medal_counts = medal_counts.fillna(0).applymap(int)
medal_counts.head()
```

```python
medal_counts.tail()
```

```python
medal_counts.info()
```
